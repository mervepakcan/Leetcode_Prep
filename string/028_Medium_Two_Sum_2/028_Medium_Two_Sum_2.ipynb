**SOLUTION**

```python
from typing import List

class Solution:
    def twoSum(self, numbers: List[int], target: int) -> List[int]:
        # initialize two pointers: left (l) at start, right (r) at end
        l, r = 0, len(numbers) - 1

        # move the pointers until they meet
        while l < r:
            curSum = numbers[l] + numbers[r]  # calculate the current sum of two numbers

            if curSum > target:
                # if the sum is too large, move the right pointer left to make it smaller
                r -= 1
            elif curSum < target:
                # if the sum is too small, move the left pointer right to make it larger
                l += 1
            else:
                # if the sum equals the target, return positions, +1 added(started from 1 index at question)
                return [l + 1, r + 1]

        # return empty list if no valid pair
        return []
```

**Time Complexity O(n)**	Single pass with two pointers

**Space Complexity	O(1)**	Uses only constant extra variables

```python
solution = Solution()

# Example 1
print(solution.twoSum([2, 7, 11, 15], 9))   # Output: [1, 2]

# Example 2
print(solution.twoSum([2, 3, 4], 6))        # Output: [1, 3]

# Example 3
print(solution.twoSum([-1, 0], -1))         # Output: [1, 2]
```
