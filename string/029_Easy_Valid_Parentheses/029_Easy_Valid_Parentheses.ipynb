**SOLUTION 1: Brute Force**


```python
class Solution:
    def isValid(self, s: str) -> bool:
        # Keep removing valid pairs until no more exist
        while '()' in s or '{}' in s or '[]' in s:
            s = s.replace('()', '')   # remove all '()' pairs
            s = s.replace('{}', '')   # remove all '{}' pairs
            s = s.replace('[]', '')   # remove all '[]' pairs

        # If string is empty, all parentheses were valid and removed
        return s == ''
```

**Time	O(n²)**	Each replace scans 0(n) and recreates the entire string multiple times O(n)

**Space	O(n)**	New string copies are created for every replace call

**SOLUTION 2: STACK**

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        # stack will hold opening brackets that haven’t been matched yet

        closeToOpen = { ")" : "(", "]" : "[", "}" : "{" }
        # dictionary (hash map) mapping each closing bracket to its matching opening bracket

        for c in s:  # iterate over every character in the string
            if c in closeToOpen:
                # if the character is a closing bracket
                if stack and stack[-1] == closeToOpen[c]:
                    # check if the top of the stack has the matching opening bracket
                    stack.pop()  # if yes → remove it (pair matched successfully)
                else:
                    # if stack is empty or top doesn't match → invalid string
                    return False
            else:
                # if the character is an opening bracket → push it onto the stack
                stack.append(c)

        # if the stack is empty at the end → all brackets matched correctly
        # if not empty → some opening brackets were never closed
        return True if not stack else False
```

**Time	O(n)**	Each character is checked once, and each operation is O(1)

**Space	O(n)**	In the worst case, all characters may be stored in the stack
