**SOLUTION 1: Reverse String**

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        newStr = ''  # store cleaned (alphanumeric & lowercase) string
        for c in s:
            if c.isalnum():        # keep only letters and digits
                newStr += c.lower() # convert to lowercase
        # compare string with its reverse
        return newStr == newStr[::-1]
```

**Time Complexity: O(n)**
Going through each character once to build the cleaned string → linear time.

**Space Complexity: O(n)**
Creating a new string (newStr) that stores all alphanumeric characters → proportional to input size.


**SOLUTION 2: Two Pointers**

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        l, r = 0, len(s) - 1  # initialize two pointers

        while l < r:
            # move left pointer if current char is not alphanumeric
            while l < r and not self.alphaNum(s[l]):
                l += 1
            # move right pointer if current char is not alphanumeric
            while r > l and not self.alphaNum(s[r]):
                r -= 1
            # compare characters (case-insensitive)
            if s[l].lower() != s[r].lower():
                return False
            # move both pointers toward the center
            l, r = l + 1, r - 1

        # all characters matched
        return True

    def alphaNum(self, c):
        # check if character is a letter or digit using ASCII values
        return (ord('A') <= ord(c) <= ord('Z') or
                ord('a') <= ord(c) <= ord('z') or
                ord('0') <= ord(c) <= ord('9'))
```

**Time Complexity: O(n)**
The algorithm scans the entire string once using two pointers (l and r), so each character is checked at most one time.

**Space Complexity: O(1)**
Only a few variables (l, r) are used, and no additional data structures are created — constant extra space.
