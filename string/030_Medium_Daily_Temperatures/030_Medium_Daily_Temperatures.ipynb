**SOLUTION: MONOTONIC STACK**

**Goal:**
Given daily temperatures, find how many days you must wait until a warmer day.
If no warmer day comes, return 0 for that day.

**Key Idea:**
Use a monotonic decreasing stack to keep track of indices of days with unresolved (yet-to-find-warmer) temperatures.

When a new temperature is higher than the temperature at the top of the stack, it means that day has found a warmer future day â€” we pop it and calculate the number of days waited.

```python
class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        st = []  # stack to store indices of days (which day it is ,0,1,2)
        res = [0] * len(temperatures)  # result array initialized with zeros

        for i in range(len(temperatures)):
            # While current day is warmer than the last stacked day
            while st and temperatures[i] > temperatures[st[-1]]:
                idx = st.pop()                # pop the previous cooler day's index and save as idx
                res[idx] = i - idx            # calculate how many days it took to get warmer
            st.append(i)                      # push current day's index to stack

        return res  # each index holds days to wait for a warmer temp (0 if none)
```

**Time Complexity: O(n)**
Each index is pushed and popped from the stack at most once,
so all operations together take linear time.

**Space Complexity: O(n)**
The stack and result array can each hold up to n elements.
