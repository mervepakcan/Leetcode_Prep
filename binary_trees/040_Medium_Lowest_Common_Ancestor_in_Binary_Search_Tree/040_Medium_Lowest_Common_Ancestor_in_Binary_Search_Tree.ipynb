The problem asks you to find the lowest common ancestor (LCA) of two nodes p and q in a binary search tree.
The LCA is the lowest (deepest) node in the tree that has both p and q as descendants, and a node is allowed to be a descendant of itself.
In short: find the nearest shared ancestor of p and q in the BST.

**SOLUTION 1 : ITERATION**

```python
class Solution:
    def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
        # Start searching from the root node
        cur = root

        while cur:
            # If both p and q are greater than the current node,
            # then the LCA must be in the right subtree
            if p.val > cur.val and q.val > cur.val:
                cur = cur.right

            # If both p and q are smaller than the current node,
            # then the LCA must be in the left subtree
            elif p.val < cur.val and q.val < cur.val:
                cur = cur.left

            # Otherwise, the current node splits p and q
            # which means the current node is the LCA
            else:
                return cur
```

**The time complexity is O(h)** because the algorithm follows a single path down the BST, where h is the treeâ€™s height.

At each step, it moves either left or right without revisiting nodes, so the work is proportional only to the height.

**The space complexity** is also O(h) in the recursive version due to the call stack
