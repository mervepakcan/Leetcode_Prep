Remove the n-th node from the end of a singly linked list and return the modified list.
The goal is to complete this in a single traversal using the two-pointer technique.
A dummy node simplifies edge cases such as removing the first element.

**SOLUTION: TWO POINTER**

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        # Create a dummy node that points to the head.
        # This helps handle edge cases cleanly, such as removing the first node.
        dummy = ListNode(0, head)

        # Initialize two pointers:
        # 'left' starts at the dummy node, and 'right' starts at the head.
        left = dummy
        right = head

        # Move 'right' pointer n steps ahead to create a gap of n between 'left' and 'right'.
        while n > 0:
            right = right.next
            n -= 1

        # Move both pointers one step at a time until 'right' reaches the end.
        # After this loop, 'left' will be pointing to the node *before* the one we need to remove.
        while right:
            left = left.next
            right = right.next

        # Remove the target node by skipping it in the linked list.
        left.next = left.next.next

        # Return the new head (which might be different if the first node was removed).
        return dummy.next
```

**Time Complexity: O(n)** The list is traversed once, where n is the number of nodes.

**Space Complexity: O(1)** Uses only a few pointers, requiring constant extra space.
