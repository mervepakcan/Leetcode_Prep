**SOLUTION:Reverse And Merge**

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next

class Solution:
    def reorderList(self, head: Optional[ListNode]) -> None:
        # Step 1: Find the middle of the linked list
        # Use slow & fast pointers:
        # - slow moves 1 step
        # - fast moves 2 steps
        # When fast reaches the end, slow will be at the middle
        slow, fast = head, head.next
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
        # Example: For 1→2→3→4→5, slow ends at 3
        # First half: 1→2→3, Second half starts at 4

        # Step 2: Reverse the second half
        # Disconnect the first half (slow.next = None)
        # 'second' starts from slow.next (the beginning of second half)
        # 'prev' will be used to reverse the links
        second = slow.next
        prev = slow.next = None  # also breaks 3→4 connection (splitting list)
        while second:
            tmp = second.next    # temporarily store next node
            second.next = prev   # reverse pointer direction
            prev = second        # move prev forward
            second = tmp         # move second forward
        # After this loop, second half is reversed
        # e.g., 4→5 becomes 5→4
        # 'prev' now points to 5 (the new head of reversed half)

        # Step 3: Merge both halves alternately
        # Now we have:
        # first = 1→2→3
        # second = 5→4
        # We will merge them like 1→5→2→4→3
        first, second = head, prev
        while second:
            # Save next nodes before overwriting
            tmp1, tmp2 = first.next, second.next

            # Connect one node from first, then one from second
            first.next = second
            second.next = tmp1

            # Move both pointers forward to continue merging
            first, second = tmp1, tmp2
        # Loop stops when second half ends
        # Result: 1→5→2→4→3
```

**Time Complexity: O(n)** each node is visited once

**Space Complexity: O(1)** in-place modification using pointers only
