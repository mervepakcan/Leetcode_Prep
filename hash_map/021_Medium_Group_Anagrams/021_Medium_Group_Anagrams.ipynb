SOLUTION W/ HASH MAP

```python
from collections import defaultdict

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        # Create a map where the key is the sorted letters and the value is a list of words
        anagram_map = defaultdict(list)
        result = []

        for s in strs:
            # Sort each word to form the key (e.g., "eat" -> "aet")
            sorted_s = tuple(sorted(s))
            # Add the original word to the list of this key
            anagram_map[sorted_s].append(s)

        # Collect all grouped anagrams into the result list
        for value in anagram_map.values():
            result.append(value)

        return result
```

defaultdict = a dictionary that automatically creates a default value (like an empty list) for new keys — so you don’t get a KeyError.

Used tuple() because dictionary keys must be immutable (unchangeable).

Time complexity is **O(n * k log k)** because we sort each of the n strings (each of length k) before grouping them. Sorting a string takes O(k log k) time.
The space complexity is **O(n * k)** because we store all words (and their sorted keys) in memory.

SOLUTION 2

```python
from collections import defaultdict

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        anagram_map = defaultdict(list)

        for s in strs:
            # Create a list of 26 zeros (for 26 lowercase letters)
            count = [0] * 26  # 26 letters
            for ch in s:
                # Use ASCII values to find the index (0 for 'a', 1 for 'b', ..., 25 for 'z')
                count[ord(ch) - ord('a')] += 1
            # Convert list to tuple (because list can't be a dict key) and add the word
            anagram_map[tuple(count)].append(s)

        return anagram_map.values()
```

TIME COMPLEXITY: O(n * k)
n = number of words, k = average length of each word
For each of the n words, we loop through its k characters once.
We do not sort, so it's faster than O(n * k log k).

SPACE COMPLEXITY: O(n * k)
Because we store all words in the dictionary and each key (count array) uses O(26) space (constant).
So overall proportional to total input size.
