Solution 1: Sorting

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        # Sort the list so duplicates will be next to each other
        nums.sort()

        # Compare each element with the previous one
        for i in range(1, len(nums)):
            if nums[i] == nums[i - 1]:
                return True   # Found a duplicate

        return False          # No duplicates found


#  Time Complexity: O(n log n)
# Because sorting the array takes O(n log n), and the single loop is O(n), which is smaller â€” total O(n log n).

# Space Complexity: O(1)
# Sorting is done in place (no extra data structures used).
```

SOLUTION 2: Set

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        # Create an empty set to store unique numbers
        num_set = set()

        # Loop through each number in the list
        for n in nums:
            # If number already in set â†’ duplicate found
            if n in num_set:
                return True
            # Otherwise, add number to the set
            num_set.add(n)

        # If loop finishes, there are no duplicates
        return False


# Time Complexity: O(n)
# We check each number once, and both "in" and "add" operations in a set take O(1) on average.

# Space Complexity: O(n)
# In the worst case (no duplicates), we store all n numbers in the set.
```

Solution 3

```python
class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        # Convert list to a set (removes duplicates)
        # If set length < list length â†’ there was a duplicate
        return True if len(set(nums)) < len(nums) else False


# ðŸ•’ Time Complexity: O(n)
# Because creating a set() processes each element once.

# ðŸ’¾ Space Complexity: O(n)
# The set stores up to all elements in memory.

```
