**SOLUTION W/ HASH MAP**

```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        sub_num = {0: 1}          # Stores prefix sums and their frequencies
        total = count = 0         # total = running sum, count = number of valid subarrays

        for n in nums:
            total += n            # Update running sum

            # If (total - k) seen before, subarrays ending here sum to k
            if total - k in sub_num:
                count += sub_num[total - k]

            # Record current prefix sum frequency
            sub_num[total] = 1 + sub_num.get(total, 0)

        return count              # Return total number of subarrays summing to k
```

**Time Complexity: O(n)** → Each element is processed once, and all hashmap operations run in constant time.

**Space Complexity: O(n)** → The hashmap stores up to n different prefix sums.
