**SOLUTION 1:Brute force**

```python
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        res = 0

        # Try every possible pair of buy and sell days
        for i in range(len(prices)):
            buy = prices[i]

            # For each buy day, try selling on every future day
            for j in range(i + 1, len(prices)):
                sell = prices[j]

                # Update the maximum profit if this pair gives a better result
                res = max(res, sell - buy)

        # Return the best profit found
        return res
```

**Time Complexity: O(n²)**
Because we use two nested loops and check every possible pair of buy and sell days, resulting in n × n operations in the worst case.

**Space Complexity: O(1)**
We only use a few variables (res, buy, sell) and do not allocate any extra data structures, so the memory usage is constant.


**SOLUTION 2: Greedy / Single-pass algorithm**

```python
from typing import List

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        buy_price = prices[0]  # cheapest so far
        profit = 0             # max profit

        for p in prices[1:]:   # start from day 2
            if p < buy_price:
                buy_price = p  # update cheapest
            profit = max(profit, p - buy_price)  # update max profit

        return profit

```

**Time Complexity: O(n)**
We only scan the list once, updating the minimum buy price and the maximum profit in a single pass.

**Space Complexity: O(1)**
We use only a few variables (buy_price and profit), so memory usage stays constant regardless of input size.

```python
# Example tests
solution = Solution()

prices1 = [7,1,5,3,6,4]
print(solution.maxProfit(prices1))


prices2 = [7,6,4,3,1]
print(solution.maxProfit(prices2))
```
